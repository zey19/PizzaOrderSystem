```python
class Pizza:
  def get_cost(self):
    return self.cost
  
  def get_description(self):
    return self.description
```

```python
class Hawaiian(Pizza):
  description = " Hawaiian Pizza : pizza sauce, cheese, cooked ham, and pineapple"
  cost = 15
   

class Margherita(Pizza):
  description = "Margherita Pizza: tomatoes, mozzarella cheese, garlic, fresh basil, and extra-virgin olive oil "
  cost = 10

class TurkishPizza(Pizza):
  description = "Turkish Pizza: beef or lamb, fresh mint, red onions, crumbled cheese"
  cost = 20
 

class Classic(Pizza):
  description = "Classic Pizza: fresh marinara sauce, mozzarella cheese, and basil"
  cost = 25
 
```

```python
class Decorator(Pizza):
  


  def get_cost(self):
       pass


  def get_description(self):
       return self.component.get_description() + \
         ' ' + Pizza.get_description(self)
```

```python
class Olives(Decorator):
  
  description="olives"
  cost = 2
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
  
  
     

class GoatCheese(Decorator):
  description = "goat cheese "
  cost = 1
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
  

class Mushrooms(Decorator):
  description = "mushrooms"
  cost = 5
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
 

class Beef(Decorator):
  description = "beef"
  cost = 7
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
  

class Onions(Decorator):
  description = "onions"
  cost = 3
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
  

class Corns(Decorator):
  description = "corns"
  cost = 3
  def __init__(self, pizza):
        self.pizza = pizza
  def get_description(self):
       return  self.pizza.get_description() + \
       ' + your chosen sauce:' + self.description
  def get_cost(self):
       return self.pizza.get_cost() + \
         self.cost
  
```

```python
def addSauce(p):
  print("Please select your sauce!")
  selectedSauce = int(input())
  if selectedSauce ==11:
    p = Olives(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  elif selectedSauce == 12:
    p = GoatCheese(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  elif selectedSauce == 13:
    p = Mushrooms(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  elif selectedSauce == 14:
    p = Beef(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  elif selectedSauce == 15:
    p = Onions(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  elif selectedSauce == 16:
    p = Corns(p)
    print("Description of your special pizza: " , p.get_description())
    print("Total: ", p.get_cost())
  return p 

```

```python

def main():
  print("Welcome to the Pizza Shop!!")
  print("***************************")
  print("Please enter your Name:")
  name = input()
  
  menu_display()
  print("Please select your pizza" ,name)
  selectedPizza = int(input())
  if selectedPizza == 1:
    p = Classic()
  elif selectedPizza == 2:
    p = Margherita()
  elif selectedPizza == 3:
    p = TurkishPizza()
  elif selectedPizza == 4:
    p = Hawaiian()
  print("Do you want to add sauce? (y/n)")
  answer = input()
  while(answer == 'y'):
    p = addSauce(p)
    print("Do you want to add sauce? (y/n)")
    answer = input()
  print("Your total fee: ", p.get_cost(), "$ Bon Appétit")
  print("For your order please enter your ID:")
  id = int(input())
  print("Please enter your credit card number and your password:")
  credit_card= int(input())
  pw = int(input())

  x = datetime.datetime.now()

  header = ['Name', 'ID', 'Order', 'Fee', 'Date', 'CreditCard Number', 'CreditCard Password']
  data = [name , id, p.get_description(), p.get_cost(), x.strftime("%x"), credit_card, pw]


  with open('order.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    # write the data
    writer.writerow(data)


  
    

if __name__ == "__main__":
    main()
```
